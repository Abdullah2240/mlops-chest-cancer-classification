```python
import mlflow
import os
from dotenv import load_dotenv

load_dotenv(override=True)
tracking_uri = os.getenv("DAGSHUB_TRACKING_URI")
token = os.getenv("DAGSHUB_TOKEN")
print(tracking_uri)
print(token)

mlflow.set_tracking_uri(tracking_uri)
```

```python
print("Current working directory:", os.getcwd())
print("Contents of current directory:", os.listdir())
```

```python
import subprocess

try:
    commit_id = subprocess.check_output(["git", "rev-parse", "HEAD"]).decode().strip()
    print("Detected Git commit ID:", commit_id)
except Exception as e:
    print("Failed to detect commit ID:", e)
```

```python
mlflow.start_run()
mlflow.log_param("commitId", 67)
mlflow.end_run()
```
